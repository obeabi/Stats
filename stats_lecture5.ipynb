```python
## Guassian distributio plot
import numpy as np
from scipy import stats
import seaborn as sns
import statsmodels.api as sm
import pylab as py
import matplotlib.pyplot as plt
```

```python
# random data points: standard normal distribution
x = np.random.normal(loc= 0.06, scale=0.97,size=100)

x
```

```python
sm.qqplot(x, line = '45')
py.show()
```

```python
sns.displot(x)
py.show()
```

```python
sns.displot(x = x, kind= 'kde')
py.show()
```

```python
# Lognormal sample
sample_size = 10000
log_mean = 0
log_sd = 1

x2 = np.random.lognormal(mean = log_mean , sigma= log_sd, size= sample_size)

sns.displot(x2)
py.show()
```

```python
sns.displot(x2, kind= 'kde')
py.show()
```

```python
sm.qqplot(x2, line= '45')
py.show()
```

```python
sm.qqplot(np.log(x2), line = '45')
py.show()
```

```python
# Generate exponetial distribution
x3 = np.random.exponential( scale = 2.2, size = 10000)
# Transform data & save lambda value
fitted_data, fitted_lambda =stats.boxcox(x3)

# plot the orginal data (non-normal) and fited data (normal)
fig,ax = plt.subplots(1,2)
sns.distplot(x3, kde = True, hist = False, kde_kws = {'shade': True, 'linewidth':2}, label = 'Non-Normal',color = 'green', ax = ax[0])

sns.distplot(fitted_data, kde = True, hist = False, kde_kws = {'shade': True, 'linewidth':2}, label = 'Normal',color = 'green', ax = ax[1])

plt.legend(loc = 'upper right')

plt.show()
```

```python
sm.qqplot(fitted_data, line = '45')
py.show()
```

```python
# pareto distribution
x4 = np.random.pareto(a= 12, size = 1000) * 10
sns.displot(x4,kind = 'kde')
py.show()
```

```python
sm.qqplot(x4, line = '45')
py.show()
```

```python
stats.probplot(x4, dist = 'norm', sparams = (12,10), plot = py)
plt.show()
```

```python
stats.probplot(x4, dist = 'pareto', sparams = (12,10), plot = py)
plt.show()
```
