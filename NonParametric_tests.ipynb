```python
# import librraies
import numpy as np
import pingouin as pg
import scipy
import pandas as pd
```

## Example 1

A new approach to prenatal care is proposed for pregnant women living in a rural community. The new program involves in-home visits during the course of pregnancy in addition to the usual or regularly scheduled visits. A pilot randomized trial with 15 pregnant women is designed to evaluate whether women who participate in the program deliver healthier babies than women receiving usual care. The outcome is the APGAR scoretext annotation indicator measured 5 minutes after birth. Recall that APGAR scores range from 0 to 10 with scores of 7 or higher considered normal (healthy), 4-6 low and 0-3 critically low. The data are shown below.

```python
# data
df ={ 'usual_care':[8,7,6,2,5,8,7,3],
     'new_program':[9,9,7,8,10,9,6, np.nan]
}

df = pd.DataFrame(df)
df.head(10)
```

```python
# pingouin library
pg.mwu(df['usual_care'], df['new_program'], alternative = 'two-sided')
```

```python
# scipy library
scipy.stats.mannwhitneyu(np.array([8,7,6,2,5,8,7,3]), np.array([9,9,7,8,10,9,6]), use_continuity= True, alternative= 'two-sided')
```

since p_value is less than the significance level (i.e.0.05), we reject the null hypothesis which is that both populations are equal

# Tests with Matched Samples
## Wilcoxon Signed Rank Test

# Example One

Consider a clinical investigation to assess the effectiveness of a new drug designed to reduce repetitive behaviors in children affected with autism. If the drug is effective, children will exhibit fewer repetitive behaviors on treatment as compared to when they are untreated. A total of 8 children with autism enroll in the study. Each child is observed by the study psychologist for a period of 3 hours both before treatment and then again after taking the new drug for 1 week. The time that each child is engaged in repetitive behavior during each 3 hour observation period is measured. Repetitive behavior is scored on a scale of 0 to 100 and scores represent the percent of the observation time in which the child is engaged in repetitive behavior. For example, a score of 0 indicates that during the entire observation period the child did not engage in repetitive behavior while a score of 100 indicates that the child was constantly engaged in repetitive behavior. The data are shown below. 

```python
# data
df = { 'child': [ 1,2,3,4,5,6,7,8] ,
       ' before treatment': [85,70,40,65,80,75,55,20] ,
      'after one week treatment': [75,50,50,40,20,65,40,25]
    
}
# A nonparametric test is appropriate to test whether there is significant improvement in repetitive behavior before versus after treatment. The hypotheses are given below.

#H0: The median difference is zero  versus
#H1: The median difference is positive α=0.05

df = pd.DataFrame(df)
df.set_index('child', inplace = True)
df.head()
```

H0: The median difference is zero  versus

H1: The median difference is positive α=0.05

```python
# pingoiun library
pg.wilcoxon(df[' before treatment'], df['after one week treatment'], alternative='greater')
```

```python
# scipy library
from scipy.stats import wilcoxon
wilcoxon(df[' before treatment'], df['after one week treatment'], alternative='greater')
```

Because the p-value = 0.027 does not exceed the level of significance (α=0.05) we  have statistically significant evidence at α =0.05 to show that there is an improvement  after treatment as compared to before.

## Tests with More than Two Independent Samples

## Example 1
A clinical study is designed to assess differences in albumin levels in adults following diets with different amounts of protein. Low protein diets are often prescribed for patients with kidney failure. Albumin is the most abundant protein in blood, and its concentration in the serum is measured in grams per deciliter (g/dL). Clinically, serum albumin concentrations are also used to assess whether patients get sufficient protein in their diets. Three diets are compared, ranging from 5% to 15% protein, and the 15% protein diet represents a typical American diet. The albumin levels of participants following each diet are shown below. Is there is a difference in serum albumin levels among subjects on the three different diets?

```python
# scipy library
#Set up hypotheses and determine level of significance.
#H0: The three population medians are equal versus
#H1: The three population medians are not all equal α=0.05
five_percent_protien=[3.1,2.6,2.9]
ten_percent_protien = [3.8,4.1,2.9,3.4,4.2] 
fifteen_percent_protien= [4.0, 5.5,5.0,4.8]
# perform krusal wallis test
from scipy.stats import kruskal
kruskal(five_percent_protien, ten_percent_protien, fifteen_percent_protien, nan_policy='propagate')
```

We  reject H0 because p_vlaue < 0.05. Therefore, we  have statistically significant evidence at α=0.05, to show that the median serum albumin levels amongst the diets is different. 

## Example 2

A personal trainer is interested in comparing the anaerobic thresholds of elite athletes. Anaerobic threshold is defined as the point at which the muscles cannot get more oxygen to sustain activity or the upper limit of aerobic exercise. It is a measure also related to maximum heart rate. The following data are anaerobic thresholds for distance runners, distance cyclists, distance swimmers and cross-country skiers.  

 Is a difference in anaerobic thresholds among the different groups of elite athletes?

```python
# pingouinlibrary
data = { ' distance runners': [185,179,192,165,174],
         'distance cyclists' :[190,209,182,178,181],
        'distance swimmers':[166,159,170, 183,160],
        'cross country skiers': [201,195,180,187,216]
}

df = pd.DataFrame(data)
df.head()
```

```python
#Set up hypotheses and determine level of significance.
#H0: The four population medians are equal versus
#H1: The four population medians are not all equal α=0.05

# perform krusal wallis test
from scipy.stats import kruskal
kruskal(df[' distance runners'], df['distance cyclists'], df['distance swimmers'],df['cross country skiers'], nan_policy='propagate')
```

e have statistically significant evidence at α =0.05, to show that there is a difference in median anaerobic thresholds among the four different groups of elite athletes.
